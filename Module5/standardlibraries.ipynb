```python
##Data serialization and desrialization

import json

data={"Name":"Sampat","Age":38}

packed_data=json.dumps(data)
print(packed_data)
print(type(packed_data))

unpacked_data=json.loads(packed_data)
print(unpacked_data)
print(type(unpacked_data))
```

```python
##CSV Reader and writer

import csv

with open("example.csv",mode='w', newline='') as file:
    writer=csv.writer(file)
    writer.writerow(["Name", "Age"])
    writer.writerow(["Sampat",38])
```

```python
##Date time

from datetime import datetime, timedelta

time_now=datetime.now()
print(time_now)

yesterday=datetime.now()-timedelta(days=1)
print(yesterday)
```

```python
##Regular expression

import re

pattern='\\d+'
sentence="I want 123 apples now"
found=re.search(pattern,sentence).group()
print(found)
```
