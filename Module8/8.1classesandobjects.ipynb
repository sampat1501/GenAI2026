```python
class Car:
    def __init__(self, name, age):
        self.name = name
        self.age = age


audi = Car("Q3", 2)
print(audi)
print(audi.name)
print(audi.age)
```

```python
class student:
    def __init__(self, name, age, course, department):
        self.name = name
        self.age = age
        self.course = course
        self.department = department

    def __str__(self):
        return f"name:{self.name}, age:{self.age},course:{self.course}, department:{self.department}"


s1 = student("Atharva", 8, "CBSE", "Mechanical")
s2 = student("Charvy", 11, "ICSE", "Computer")
print(s1)
print(s2)
```

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(
            f"{amount} amount credited successfully. "
            f"New available balance is {self.balance}"
        )

    def withdraw_amount(self, amount):
        if amount > self.balance:
            print("Insufficient balance.")
        else:
            self.balance -= amount
            print(
                f"{amount} withdrawn successfully. "
                f"Current balance is {self.balance}"
            )

    def get_balance(self):
        return self.balance


# Create an object
account = BankAccount("Sampat", 5000)

# Deposit money
account.deposit(500)

# Withdraw money
account.withdraw_amount(700)

# Print balance
print(f"Available Balance: {account.get_balance()}")
```
