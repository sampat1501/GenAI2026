```python
## Name error exception

try:
    a = b
except Exception as ex:
    print(ex)

try:
    1 / 0
except Exception as e:
    print(e)
```

```python
try:
    a = b
    10 / 0
except FileNotFoundError as e:
    print(e)
except NameError as ne:
    print(ne)
    print("Name error exception is executed")
except TypeError as t:
    print(t)
except ValueError as ve:
    print(ve)
except ZeroDivisionError as zd:
    print(zd)
except Exception as ex:
    print("Exception is executed")
    print(ex)
```

```python
## Use of try except and catch

try:
    num = int(input("Enter the valid number"))
    result = 10 / num
except ValueError:
    print("Entered value is not a number")
except ZeroDivisionError:
    print("You can't divide the number by zero")
else:
    print(f"The result is '{result}' here")
```

```python
## Use of try except and else, finally block

try:
    num = int(input("Enter the valid number"))
    result = 10 / num
except ValueError:
    print("Entered value is not a number")
except ZeroDivisionError:
    print("You can't divide the number by zero")
else:
    print(f"The result is '{result}' here")
finally:
    print("Execution is completed")
```

```python
## Demonstration: try-except-else-finally with valid input

try:
    num = int(input("Enter your number"))
    result = 10 / num
except ValueError:
    print("Entered value is not a number")
except ZeroDivisionError:
    print("You can't divide the number by zero")
else:
    print(f"The result is '{result}' here")
finally:
    print("Execution is completed")
```

```python
try:
    with open("sampat.txt", "r") as file:
        content = file.read()
        print(content)
except FileNotFoundError:
    print("File does not exist")
finally:
    print("File closed successfully")
```
