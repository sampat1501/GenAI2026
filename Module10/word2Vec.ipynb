```python
import os
import gensim.downloader as api
from gensim.models import KeyedVectors

os.environ["GENSIM_DATA_DIR"] = r"D:\GenAI2026\gensim-data"

model_name = "word2vec-google-news-300"
model_file = r"D:\GenAI2026\word2vec-google-news-300.kv"

try:
    wv = KeyedVectors.load(model_file)
    print("Loaded from local file")
except FileNotFoundError:
    wv = api.load(model_name)  # download once
    wv.save(model_file)  # save for later
    print("Downloaded and saved")
```

```python
wv.most_similar("king", topn=10)
```
