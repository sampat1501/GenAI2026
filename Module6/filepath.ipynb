```python
##Getting the absolute path
import os
path="sampat.txt"

if (os.path.exists(path)):
    print(f"The path'{path} is exist")
else:
    print(f"the path {path} is not exist")

```

```python
## Checking the path is file or directory

import os

file_path = "sample.txt"

if os.path.isfile(path):
    print(f"The path'{file_path}' is file")
elif os.path.isdir(file_path):
    print(f"The path is'{file_path}'directory")
else:
    print(f"The path '{file_path}' is niether file nor directory")
```
