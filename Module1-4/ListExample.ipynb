```python
to_do_list=["Wake up", "Brush teeth", "Eat breakfast", "Go to work", "Attend meetings", "Complete tasks", "Go home", "Relax", "Sleep"]

to_do_list.append("Go to office")
print(to_do_list)

to_do_list.remove("Wake up")
print(to_do_list)

for task in to_do_list:
    if task == "Attend meetings":
        print("Don't forget to prepare for the meeting!")

print("Today's tasks:")
for task in to_do_list:
    print("- " + task)
```

```python
student_grades=[85, 92, 78, 90, 88]
print(student_grades)

student_grades.append(95)
print(student_grades)

avg_student_grade=sum(student_grades)/len(student_grades)
print("Average student grade: ", avg_student_grade)

max_student_grade=max(student_grades)
print("Highest student grade: ", max_student_grade)

min_student_grade=min(student_grades)
print("Min student grade is: ", min_student_grade)

for grade in student_grades:
    if grade==95:
        print("Outstanding performance! Keep it up!")

student_grades.remove(78)
print(student_grades)


```
