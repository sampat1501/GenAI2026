```python
students={"student1":{"name": "Alice", "age": 20, "grade": "A"}, "student2":{"name": "Bob", "age": 22, "grade": "B"}, "student3":{"name": "Charlie", "age": 21, "grade": "A"}}

for student_id, student_info in students.items():
    print(f"Student ID: {student_id}, Student Info: {student_info}")
    for key , value in student_info.items():
        print(f"  {key}: {value}")
```

```python
## Dictionary comprehension
squares = {i: i**2 for i in range(10)}
print(squares)
```

```python
##Dictioanry comprehension with condition
multiples = {i: i*2 for i in range(10) if i%2==0}
print(multiples)
```

```python
##count occurrences of each element in a list using dictionary comprehension
lst=[1,2,2,3,3,3,4,4,4,4]
occurrences = {i: lst.count(i) for i in set(lst)}
print(occurrences)
```
