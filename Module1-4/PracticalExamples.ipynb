```python
def temp_conversion(temp, unit):
    if unit=="C":
        return temp*9/5+32
    elif unit=="F":
        return(temp-32)*5/9
    else:
        return "Invalida entity entered. Please enter 'C' for Celsius or 'F' for Fahrenheit."

ftemp=temp_conversion(100, "C")
print(ftemp)

ctemp=temp_conversion(225,"F")
print(ctemp)
```

```python
def is_strong_password(password):
    '''This Function check password is strong or not'''
    if len(password)<8:
        return False ,"lenght is less than 8"
    if not any(char.isdigit() for char in password):
        return False, "does not contain any digit "
    if not any (char.islower() for char in password):
        return False , "does not contain any lower case char "
    if not any(char.isupper() for char in password):
        return False,  "does not contain any upper case character "
    if not any(char in "@!#$%^&*" for char in password):
        return False,  "does not contain any special character "
    else:
        return True
##Call function
is_strong_password("Passtw@1")

```

```python
## calculate the total cost of items in the cart

def total_cost_items(cart):
    total_cost=0
    for item in cart:
        total_cost+=item["price"]*item["quantity"]
    return total_cost


cart=[
    {"name":"apple", "price":0.4, "quantity":4 },
    {"name":"Banana", "price":0.6, "quantity":6},
    {"name":"Papaya", "price":0.5, "quantity":5}
]

total_cost=total_cost_items(cart)
print(total_cost)
```

```python
def palindrome_string(s):
    s.lower().replace(" ","")
    return s==s[::-1]

## call Function

print(palindrome_string("a man a plan a canal panama"))
print(palindrome_string("level"))
print(palindrome_string("Hello"))

```

```python
def factorial(n):
    if n==0:
        return 1
    return n* factorial(n-1)

print(factorial(6))
```

```python
import re


def word_count_freq(file_path):
    word_count={}
    with open(file_path,"r") as file:
        for line in file:
          words=line.split()
          for word in words:
             word=word.lower().strip()
             word_count[word]=word_count.get(word,0)+1
    return word_count


## calling a function
word_count_freq("sample.txt")

```
